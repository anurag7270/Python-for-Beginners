```python
class Human:
    def __init__ (self,n,o):
        self.name=n
        self.occupation=o
    
    def do_work(self):
        if self.occupation=="actor":
            print(self.name,"shoot a film")
        elif self.occupation=="tennis player":
            print(self.name,"plays tennis")
    def speaks(self):
        print(self.name,"says how are you?")
           
```

```python
tom=Human('Tom Cruise',"actor")
tom.do_work()
tom.speaks()
```

```python
maria=Human('Maria Sharapova',"tennis player")
maria.do_work()
maria.speaks()
```

```python
class Vehicle:
    def general_usage(self):
        print("general use: transportation")

class Car(Vehicle):
    def __init__ (self):
        print("I am car")
        self.wheel=4
        self.has_roof=True
    def specific_usage(self):
        print("Specific use: commute to work, family vacation")
        
class MotorCycle(Vehicle):
    def __init__ (self):
        print("I am MotorCycle")
        self.wheel=2
        self.has_roof=False
    def specific_usage(self):
        print("Specific use: road trip, racing")
    
```

```python
c=Car()
c.general_usage()
c.specific_usage()

m=MotorCycle()
m.general_usage()
m.specific_usage()
```

```python
print(isinstance(c,MotorCycle))
```

```python
print(isinstance(c,Car))
```

```python
print(issubclass(MotorCycle,Vehicle))
```

```python
print(issubclass(Car,Vehicle))
```

```python
class Father():
    def gardening(self):
        print("I love gardening")
        
class Mother():
    def cooking(self):
        print("I love cooking")
        
class Child(Father,Mother):
    def sports(self):
        print("I enjoy Sports")
```

```python
c=Child()
c.gardening()
c.cooking()
c.sports()
```

```python
class RemoteControl():
    def __init__(self):
        self.channels=["HBO","CNN","espn"]
        self.index = -1
        
    def __iter__(self):
        return self
        
    def __next__(self):
        self.index +=1
        if self.index == len(self.channels):
            raise StopIteration
        return self.channels[self.index]
    
```

```python
r =RemoteControl()
itr = iter(r)
print(next(itr))
print(next(itr))
```
