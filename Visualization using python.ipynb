# MatplotLib

Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy.
Matplotlib is one of the most popular Python packages used for data visualization. It is a cross-platform library for making 2D plots from data in arrays. 

Some of the major Pros of Matplotlib are:

1. Generally easy to get started for simple plots
2. Support for custom labels and texts
3. Great control of every element in a figure
4. High-quality output in many formats
5. Very customizable in general

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
import numpy as np
```

```python
# Let's create a simple example using numpy
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
# plotting graph
plt.scatter(x,y,c="g")
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title("Graph")
# plt.savefig('projet_1.png')
# plt.show()
```

```python
y = x*x
plt.plot(x,y,'bo',linestyle='dashed',linewidth=2, markersize=10)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title("2D-Graph of y=x*x")
```

```python
# subplots
plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'b--')
plt.subplot(2,2,3)
plt.plot(x,y,'g*--')
plt.subplot(2,2,4)
plt.plot(x,y,'bo')
```

```python
x = np.arange(1,11)
y= 4 * x + 10
plt.plot(x,y,'ro',linestyle='dashed',linewidth=2, markersize=5)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title("2D-Graph of y=4*x+10")
```

```python
x =np.arange(0, 5*np.pi,0.1)
y = np.sin(x)
plt.title("Sine Wave")
plt.plot(x,y)
plt.show()
```

**Subplot**

```python
# subplot
x =np.arange(0, 3*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
plt.subplot(2,1,1)

plt.title("Sine Wave")
plt.plot(x,y_sin,'r--')

plt.subplot(2,1,2)
plt.title("Cos Wave")
plt.plot(x,y_cos,'b--')
```

**Bar Plot**

```python
# Bar plot
x=[2,8,10]
y=[11,16,9]

x2=[3,9,11]
y2=[6,15,7]
plt.bar(x,y,color='r')
plt.bar(x2,y2,color='b')
```

**Histogram**

```python
# Histogram
a = np.array([22,56,24,15,65,78,62,98,48,32,45,78,54,79,31,27])
plt.hist(a)
```

```python
# Box plot
x=  [np.random.normal(0, std, 100) for std in range(1,5)]

plt.boxplot(x, vert=True,patch_artist=True)
```

```python
# Pie Chart
y = 'IIT','GALGOTIYA','AMITY','NIT'
x = [210,120,255,215]
colors =['gold','green','blue','lightcoral']
explode= (0.4,0,0,0)
plt.pie(x, labels=y, explode = explode, autopct = '%1.1f%%', shadow=True)
plt.axis('equal')
```

# Seaborn

Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

There are three types of plots in seaborn:
1. Displot
2. Joinplot
3. Pairplot

```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
```

```python
df.head()
```

**Correlation with Heatmap**

Correlation matrices are an essential tool of exploratory data analysis. Correlation heatmaps contain the same information in a visually appealing way. What more: they show in a glance which variables are correlated, to what degree, in which direction, and alerts us to potential multicollinearity problems.

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

**Joinplot**

Join plot allows to study the relationship between 2 numeric variables. The central display their correlation. It is usually a scatterplot, hexbin plot, a 2D histogram or a density plot. 

**Univariate Analysis**

```python
sns.jointplot(x = "tip", y="total_bill", data = df, kind='hex')
```

```python
sns.jointplot(x = "tip", y="total_bill", data = df, kind='scatter')
```

x, y : vectors or keys in ``data``
    Variables that specify positions on the x and y axes.

data : :class:`pandas.DataFrame`, :class:`numpy.ndarray`, mapping, or sequence
    Input data structure. Either a long-form collection of vectors that can be
    assigned to named variables or a wide-form dataset that will be internally
    reshaped.

kind : { "scatter" | "kde" | "hist" | "hex" | "reg" | "resid" }
    Kind of plot to draw. See the examples for references to the underlying functions.

color : :mod:`matplotlib color <matplotlib.colors>`
    Single color specification for when hue mapping is not used. Otherwise, the
    plot will try to hook into the matplotlib property cycle.

**Pair Plot**

A pairs plot allows us to see both distribution of single variables and relationships between two variables. Pair plots are a great method to identify trends for follow-up analysis and, fortunately, are easily implemented in Python!

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

**Distplot**

 distplot plots a univariate distribution of observations. The distplot () function combines the matplotlib hist function with the seaborn kdeplot () and rugplot () functions. 

```python
sns.displot(df)
```

```python
sns.distplot(df['total_bill'])
```

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=5,color='r')
```

**Box plot**

Boxplots are a measure of how well distributed the data in a data set is. It divides the data set into three quartiles. This graph represents the minimum, maximum, median, first quartile and third quartile in the data set. It is also useful in comparing the distribution of data across data sets by drawing boxplots for each of them.

```python
sns.boxplot('smoker','total_bill',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='h')
```

```python
sns.boxplot(x="total_bill", y="day", hue="smoker",data=df)
```

**Violin plot**

Violin plots are similar to box plots, except that they also show the probability density of the data at different values. 

```python
sns.violinplot(x='total_bill',y='day',data=df,palette='rainbow')
```
